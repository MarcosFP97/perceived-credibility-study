##### General Conclusions



*   Users are generally good at identifying medium to highly credible documents, but they struggle with low quality ones.
*   They tend to overscore the low quality ones => motivates the need of doing more reseaarch on empowering users with more sophisticated means for judging credibility.
*   "It seems" (proof) that our guidelines (objective) correlate up to a certain point with user credibility judgements

##### Per topic
**T1**: Do antioxidants help female subfertility?

**T5**: Do sealants prevent dental decay in the permanent teeth?

**T8**: Does melatonin help treat and prevent jet lag?

**T10**: Does traction help low back pain?

*   T5 (dental decay) seems the topic that users have clearer. Meanwhile, T1 seems the most difficult one.
*   In T10 (back pain) they struggle to differentiate high levels of credibility, and in T8 (jet lag) they have problems differentiating no credible docs from medium credible.

##### Deviation


*   En el T1, los usuarios tienden a subpuntuar la credibilidad.
*   The T5 has the median almost in 0, this reinforces the idea that this was the topic that users had clearer.
*   Topic T10 they overestimate credibility. This can be due to the fact that they assign very high punctuations to '1' pages. The same for T8, but less.

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from typing import List

def get_keyterms(corpus:List[str]) -> List[str]:
  corpus = [x for x in corpus if str(x)!='nan']
  print("# of Just for category", len(corpus))
  vectorizer = TfidfVectorizer(min_df=2, stop_words='english', ngram_range=(1,3))
  X = vectorizer.fit_transform(corpus)
  df = pd.DataFrame(X.toarray())
  df.columns = vectorizer.get_feature_names_out()
  df = df.sum(axis=0).reset_index().sort_values(by=0, ascending=False)
  return list(df["index"].values[:15])
```
